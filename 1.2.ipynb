```python
user_jobid = int(input("Enter the jobid: "))

job_row = job_data.loc[job_data['Id'] == user_jobid]

jobId = job_row['Id'].values[0]

task_row = task_data.loc[task_data['jobId'] == jobId]
taskId = task_row['Id'].values[0]  

datetimeFrom = job_row['datetimeFrom'].values[0] 
datetimeTo = job_row['datetimeTo'].values[0]  

datetimeFrom = datetime.strptime(datetimeFrom, "%Y-%m-%d %H:%M:%S.%f")
datetimeTo = datetime.strptime(datetimeTo, "%Y-%m-%d %H:%M:%S.%f")

datetimeFrom_date = datetimeFrom.strftime('%Y-%m-%d')
datetimeTo_date = datetimeTo.strftime('%Y-%m-%d')

fuel_price_data['datetime'] = pd.to_datetime(fuel_price_data['datetime'], format="%Y-%m-%d %H:%M:%S.%f")
fuel_price_data['date'] = fuel_price_data['datetime'].dt.strftime('%Y-%m-%d')

user_fuel_price_data = fuel_price_data[(fuel_price_data['datetime'] >= datetimeFrom) & 
                                       (fuel_price_data['datetime'] <= datetimeTo)]

user_fuel_price_data = user_fuel_price_data[user_fuel_price_data['taskId'] == taskId]

user_fuel_price_data_group = user_fuel_price_data.groupby('fuelId').agg(
    mean_value=('value', 'mean')
)

user_fuel_price_data_group['datetimeFrom_date'] = datetimeFrom_date
user_fuel_price_data_group['datetimeTo_date'] = datetimeTo_date
user_fuel_price_data_group['taskId'] = taskId
user_fuel_price_data_group['jobId'] = jobId

fuel_data_renamed = fuel_data.rename(columns={'Id': 'fuelId', 'name': 'fuel_name'})

user_fuel_price_data_group = user_fuel_price_data_group.merge(fuel_data_renamed[['fuelId', 'fuel_name']], 
                                                              on='fuelId', 
                                                              how='left')

user_fuel_price_data_group = user_fuel_price_data_group.drop(columns=['fuelId'])

user_fuel_price_data_group.to_excel('user_fuel_price_data_group.xlsx', index=False)

user_fuel_price_data_group
```

```python
def process_fuel_price_data(job_data, task_data, fuel_price_data, fuel_data, output_file):
    with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
        for user_jobid in job_data['Id'].unique():
            job_row = job_data.loc[job_data['Id'] == user_jobid]
            jobId = job_row['Id'].values[0]
            
            task_row = task_data.loc[task_data['jobId'] == jobId]
            taskId = task_row['Id'].values[0]
            
            datetimeFrom = job_row['datetimeFrom'].values[0]
            datetimeTo = job_row['datetimeTo'].values[0]
            datetimeFrom = datetime.strptime(datetimeFrom, "%Y-%m-%d %H:%M:%S.%f")
            datetimeTo = datetime.strptime(datetimeTo, "%Y-%m-%d %H:%M:%S.%f")

            datetimeFrom_date = datetimeFrom.strftime('%Y-%m-%d')
            datetimeTo_date = datetimeTo.strftime('%Y-%m-%d')

            fuel_price_data['datetime'] = pd.to_datetime(fuel_price_data['datetime'], format="%Y-%m-%d %H:%M:%S.%f")
            fuel_price_data['date'] = fuel_price_data['datetime'].dt.strftime('%Y-%m-%d')
            
            user_fuel_price_data = fuel_price_data[(fuel_price_data['datetime'] >= datetimeFrom) & 
                                                   (fuel_price_data['datetime'] <= datetimeTo)]
            
            user_fuel_price_data = user_fuel_price_data[user_fuel_price_data['taskId'] == taskId]
            
            user_fuel_price_data_group = user_fuel_price_data.groupby('fuelId').agg(
                mean_value=('value', 'mean')
            )
            
            user_fuel_price_data_group['datetimeFrom_date'] = datetimeFrom_date
            user_fuel_price_data_group['datetimeTo_date'] = datetimeTo_date
            user_fuel_price_data_group['taskId'] = taskId
            user_fuel_price_data_group['jobId'] = jobId
            
            fuel_data_renamed = fuel_data.rename(columns={'Id': 'fuelId', 'name': 'fuel_name'})
            user_fuel_price_data_group = user_fuel_price_data_group.merge(fuel_data_renamed[['fuelId', 'fuel_name']], 
                                                                          on='fuelId', 
                                                                          how='left')
            
            user_fuel_price_data_group = user_fuel_price_data_group.drop(columns=['fuelId'])

            sheet_name = f'Job_{jobId}'  # Sheet name for each jobId
            user_fuel_price_data_group.to_excel(writer, sheet_name=sheet_name, index=False)
        
        writer.save()


output_file = 'user_fuel_price_data_all_jobs.xlsx'

process_fuel_price_data(job_data, task_data, fuel_price_data, fuel_data, output_file)

print(f"Data successfully written to {output_file}")
```
